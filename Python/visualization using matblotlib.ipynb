```python
import matplotlib.pyplot as plt
months=['Jan','Feb','Mar','Apr','May','Jun']
energy_consumption=[1200,1300,1100,1500,1400,1600]
plt.plot(months,energy_consumption)
plt.title("energy consumption over 6months")
plt.xlabel("month")
plt.ylabel("energy consumption (MWh)")
plt.show()
```

```python
energy_sources=['Solar','Wind','Hydropower','Biomass']
energy_values=[1200,3400,2900,2500]
plt.bar(energy_sources,energy_values,color='lightblue',edgecolor='black',linewidth=2)
plt.title("energy consumption by Renewable Energy Source")
plt.xlabel("Energy source")
plt.ylabel("energy consumption (MWh)")
plt.show()
```

```python
plt.pie(energy_values,labels=energy_sources,autopct='%1.1f%%')
plt.title("energy consumption share by source")
plt.show()
```

```python
carbon_emissions=[400,500,450,300,350,550]
plt.scatter(energy_consumption,carbon_emissions,color='purple')
plt.title("energy consumption vs energy emissions")
plt.xlabel("energy consumption (MWh)")
plt.ylabel("carbon emission (kg CO2)")
plt.show()
```
